```python
#importing libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
#read the csv
df = pd.read_csv("Billionaire.csv")
df
```

```python
#checking the value if there is null
df.isnull()
```

```python
#information of the csv file
df.info()
```

```python
#checking dataset at different percentile
df.describe()
```

```python
df["Country"].value_counts().head(10)
#usa have highest number of the billionaire
```

```python
df["Industry"].value_counts()
#industry in finance and investment has highest number of the billionaire
```

```python
#remove each string from 
df['NetWorth_Billion_Dollars'] = df['NetWorth'].replace({'\$': '', 'B': ''}, regex=True).astype(float)
```

```python
df
```

```python
df1 = df.drop(columns="NetWorth")
```

```python
df1.describe()
```

```python
df1.Rank.value_counts()
# means they share same rank
```

```python
####We can have a an absolute unique rank for each person
df1['Abs_rank']=df.index+1
df1
```

```python
len(df1)
# there are 2755 billionaire in the world(feb -2021)
```

```python
#### plottting networth vs rank.. we have plotted matplotlib for it above 
#only taking top 200 bilionaire
#Exploring Net Worth Vs Ranks(top 200)

plt.figure(figsize=(14,7))
plt.plot(df1.Rank[:200],df1.NetWorth_Billion_Dollars[:200],c='pink',lw=4)
plt.xlabel('Ranks')
plt.ylabel('Net Worth(Billion-$)')
plt.title('Wealth Distrbution')

sns.set_style('whitegrid')
```

### this shows as we move up the rank there is large change  in net worth

```python
#Frequency of Top 200 Billionaires Across Range of NetWorth
plt.hist(df1.NetWorth_Billion_Dollars[df1.Rank<=200],bins=10)
plt.xlabel('Range')
plt.ylabel('Frequency')


plt.title('Frequency of Top 200 Billionaires Acc. to Net Worth');
```

### as expected the count of billionaire are very low at lower range of Net Worth

```python
### top 5 domains with the most number of billionaires:

a = df1["Source"].value_counts().head()
index = a.index
sources = a.values
custom_colors = ["blue", "yellow", 'grey', "green", "red"]
plt.figure(figsize=(8, 8))
plt.pie(sources, labels=index, colors=custom_colors)
central_circle = plt.Circle((0,0), 0.5, color='white')
fig = plt.gcf()
fig.gca().add_artist(central_circle)
plt.rc('font', size=20)
plt.title("Top 5 Domains to Become a Billionaire", fontsize=20)
plt.show()
```

### In this we found that  pharmaceuticals and real estate domains have most number of billionares...

```python
###top 5 industries with the most number of billionaires:


a = df1["Industry"].value_counts().head()
index = a.index
industries = a.values
custom_colors = ["blue", "yellow", 'grey', "green", "red"]
plt.figure(figsize=(8,8))
plt.pie(industries, labels=index, colors=custom_colors)
central_circle = plt.Circle((0, 0), 0.5, color='white')
fig = plt.gcf()
fig.gca().add_artist(central_circle)
plt.rc('font', size=20)
plt.title("Top 5 Industries with Most Number of Billionaires", fontsize=20)
plt.show()
```

### In this we found that TECHNOLOGY , FINANCE & INVESTEMENTS and MANUFACTURING have most number of billionares...

```python
### top 5 countries with the most number of billionaires:

a = df1["Country"].value_counts().head()
index = a.index
Countries = a.values
custom_colors = ["blue", "yellow", 'grey', "green", "red"]
plt.figure(figsize=(8,8))
plt.pie(Countries, labels=index, colors=custom_colors)
central_circle = plt.Circle((0, 0), 0.5, color='white')
fig = plt.gcf()
fig.gca().add_artist(central_circle)
plt.rc('font', size=20)
plt.title("Top 5 Countries with Most Number of Billionaires", fontsize=20)
plt.show()
```

### At the end of the analysis of global billionaires, I found that China and the United States are the countries with the most billionaires which concludes that the business environment and the success rate of a startup is much better in the US and China than in the rest of the world.
